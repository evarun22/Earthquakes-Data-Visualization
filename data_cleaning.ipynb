```python
from geopy.geocoders import Nominatim
geolocator = Nominatim()
```

```python
location = geolocator.reverse("48.8588443, 2.2943506")

location.address.split(',')[-1]
```

```python
df["Coordinates"] = df["Latitude"].astype(str) +', '+ df["Longitude"].astype(str)
```

```python
df.head()
```

```python
df['Coordinates']=df['Coordinates'].astype(str)
```

```python
df['Coordinates'].dtype
```

```python
df['Coordinates'].head(1)
```

```python
country=[]
coordinates=df['Coordinates'].tolist()
for i in coordinates:
    a = geolocator.reverse(i)
    #print(a)
    if a.address is None:
        country.append('None')
    else:
        country.append(a.address.split(',')[-1].strip())
len(country)
```

```python
df['Country']=country
```

```python
df.head()
```

```python
df.to_csv('earthquakes_with_countries.csv')
```
